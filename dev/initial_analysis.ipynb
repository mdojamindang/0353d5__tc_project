```python
for i in df['restaurant_names'].unique():
    # print(i)
    temp = df[df['restaurant_names']==i]
    most_freq = temp['food_names'].mode().iloc[0]
    print(f"most popular for {i} is {most_freq}")
```

```python
for i in df['restaurant_names'].unique():
    # print(i)
    temp = df[df['restaurant_names']==i]
    most_visit = temp['first_name'].mode()
    most_freq = list(temp['food_names'].mode())
    if len(most_freq) == 1:
        print(f"most popular for {i} is {most_freq[0]}")
        print(f"most frequent visitor for {i} is {most_visit[0]}")
    else:
        print(f"most popular for {i} is {most_freq}")
        print(f"most frequent visitor for {i} is {most_visit}")
    grouped = temp.groupby("food_names",as_index=False)['food_cost'].sum()
    print(grouped.loc[grouped["food_cost"].idxmax()])
    print("---------------------")
    
```
